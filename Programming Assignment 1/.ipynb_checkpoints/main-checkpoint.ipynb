```python
import pandas as pd
import numpy as np
import xlrd
import random
import matplotlib.pyplot as plt
```

### Read in the concrete data and split into training and testing dataset

```python
data = pd.read_excel("data/Concrete_Data.xls")

features = data.columns
training = np.array(data[:900])
testing = np.array(data[900:])
X_train, y_train = training[:, :-1], training[:, -1]
X_test, y_test = testing[:, :-1], testing[:, -1]
```

## A) Uni-variate linear regression

For a univariate linear model: f(x) = mx + b

```python
def uni_loss_function(x, y, m, b, n):
    return np.sum((y - (m*x + b)) ** 2) / n
```

```python
def univariateLinearRegression(feature, x_train, y_train, x_test, y_test):
    n_train = x_train.shape[0]
    n_test = x_test.shape[0]
    m = 0
    b = 0
    prev_mse = float("inf")
    iteration = 0
    mse = uni_loss_function(x_train, y_train, m, b, n_train)
    while prev_mse - mse > 0.00001:
        y_pred = m * x_train + b
        m = m - 0.000000001 * (-2/n_train * np.sum(x_train * (y_train - y_pred)))
        b = b - 0.000000001 * (-2/n_train * np.sum(y_train - y_pred))
        prev_mse = mse
        mse = uni_loss_function(x_train, y_train, m, b, n_train)
        iteration += 1
    print('%s' %(feature))
    print('Coefficient m: %s' %(round(m, 3)))
    print(f"Training loss: { uni_loss_function(x_train, y_train, m, b, n_train) }")
    print(f"Testing loss: { uni_loss_function(x_test, y_test, m, b, n_test) }")
    return mse, iteration, m, b
```

```python
def plot(feature, x_train, y_train, x_test, y_test):
    mse, iteration, m, b = univariateLinearRegression(feature, x_train, y_train, x_test, y_test)
    plt.figure()
    plt.scatter(x_train, y_train, s=5)
    x_range = np.arange(np.min(x_train), np.max(x_train), 0.1)
    y_range = m * x_range + b
    plt.plot(x_range, y_range, c='red')
    plt.title(feature)
    plt.xlabel(feature)
    plt.ylabel(features[-1])
    plt.show()
    plt.close()
    print("--------------------------------------------------------------------")
```

```python
for i, feature in enumerate(features[:-1]):
    x_train = X_train[:, i]
    x_test = X_test[:, i]
    plot(features[i], x_train, y_train, x_test, y_test)
```

## B) Multi-variate linear regression

For a multivariate linear model: y = f(x) = (a . x)

```python
def multi_loss_function(x, y, a, n):
    return np.sum((np.matmul(x, a).reshape(-1,) - y) ** 2) * 1/n
```

```python
def multiVariateLinearRegression(x_train, y_train, x_test, y_test):
    n_train = x_train.shape[0]
    n_test = x_test.shape[0]
    x_train = np.hstack((np.ones(n_train)[:, np.newaxis], x_train))
    x_test = np.hstack((np.ones(n_test)[:, np.newaxis], x_test))
    a = np.zeros(x_train.shape[1])
    prev_mse = float("inf")
    iteration = 0
    mse = multi_loss_function(x_train, y_train, a, n_train)
    while prev_mse - mse > 0.00001:
        y_pred = np.matmul(x_train, a.reshape(-1, 1))
        iteration += 1
        a = a - 0.000000001 * (2/n_train * np.matmul(y_pred.reshape(1, -1) - y_train.reshape(1, -1), x_train).reshape(-1,))
        prev_mse = mse
        mse = multi_loss_function(x_train, y_train, a, n_train)
    
    print(f"Training loss: { multi_loss_function(x_train, y_train, a, n_train) }")
    print(f"Testing loss: { multi_loss_function(x_test, y_test, a, n_test) }")
    
    return mse, iteration, a
```

```python
multiVariateLinearRegression(X_train, y_train, X_test, y_test)
```

## C) Extra Credit I: Multi-variate polynomial regression

For a multivariate polynomial model: y = $ax^2$ + bx + c

```python
def multiVariatePolynomialRegression(x_train, y_train, x_test, y_test):
    n_train = x_train.shape[0]
    n_test = x_test.shape[0]
    x_train = np.hstack((x_train, x_train ** 2))
    x_test = np.hstack((x_test, x_test ** 2))
    a = np.zeros(x_train.shape[1])
    prev_mse = float("inf")
    iteration = 0
    mse = multi_loss_function(x_train, y_train, a, n_train)
    while prev_mse - mse > 0.000001:
        y_pred = np.matmul(x_train, a.reshape(-1, 1))
        iteration += 1
        a = a - 0.00000000000001 * (2/n_train * np.matmul(y_pred.reshape(1, -1) - y_train.reshape(1, -1), x_train).reshape(-1,))
        prev_mse = mse
        mse = multi_loss_function(x_train, y_train, a, n_train)
    
    print(f"Training loss: { multi_loss_function(x_train, y_train, a, n_train) }")
    print(f"Testing loss: { multi_loss_function(x_test, y_test, a, n_test) }")
    
    return mse, iteration, a
```

```python
multiVariatePolynomialRegression(X_train, y_train, X_test, y_test)
```
